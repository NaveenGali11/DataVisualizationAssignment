# Data Visualization Assignment

```python
#importing libraries
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [20, 5]
```

```python
import pandas as pd
```

```python
#reading the dataset
```

```python
data = pd.read_csv('GamingCompanies-Main.csv')
```

```python
#Printing the Head of dataset
display(data.head(10))
```

```python
#Scatter Plot using dataset
```

```python
plt.scatter(data.head(10)['Name'],data.head(10)['earnings'])
plt.title("Scatter Plot")
plt.xlabel('Company')
plt.ylabel('Earnings')
```

```python
#Bar Chart using the Data Set
```

```python
plt.bar(data.head(10)['Name'],data.head(10)['earnings'])
plt.title("Bar Chart")
plt.xlabel("Company")
plt.ylabel("Earnings")
```

```python
#Histogram using the DataSet
```

```python
plt.hist(data.head(50)['earnings'])
plt.title("Histogram")
```

```python
#line Plot using python on dataset
```

```python
plt.plot(data.head(15)['Name'],data.head(15)['earnings'],linestyle='dashed',color='red')
```

```python
#Double Line Plot on same Graph
plt.plot(data.head(13)['Name'],data.head(13)['OperatingMargin'],linestyle='dashed',color='red')
plt.plot(data.head(13)['Name'],data.head(13)['employeeCount'],linestyle='dashed',color='blue')
```

```python
labels = data.head(13)['Name']
sizes = data.head(13)['employeeCount']
fig1,ax1 = plt.subplots()
ax1.pie(sizes,labels = labels,autopct = '%1.1f%%',shadow=True,startangle=90)
ax1.axis('equal')
plt.show()
```
